# column renaming

```python
df = df.rename(columns={
    'Company': 'Employer',
    'Age': 'Employee_Age',
    'Salary': 'Monthly_Salary',
    'Place': 'Location',
    'Country': 'Employee_Country',
    'Gender': 'Employee_Gender'
})
print(df.columns)
```

# data cleaning

### 1.Find null value

```python
print(df.isnull().sum())
```

### 2.Filling the null values with mean values and mode values


```python
df.mean
```

```python
df = df.fillna(df.mean(numeric_only=True))
```

```python
df.isnull().sum()
```

```python
df.mode
```

```python
df['Employer'].fillna(df['Employer'].mode().iloc[0],inplace=True)
df['Location'].fillna(df['Location'].mode().iloc[0],inplace=True)
```

```python
df.isnull().sum()
```

### 3.Remove duplicates

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
df[df.duplicated()]
```

### 4.Finding the outliers(monthly_salary and employee_age)

```python
Q1 = df['Monthly_Salary'].quantile(0.25) # Calculate IQR of column Monthly_Salary
Q3 = df['Monthly_Salary'].quantile(0.75)
IQR = Q3 - Q1
print("Q1 =",Q1)
print("Q3 =",Q3)

# Identify outliers
outliers_iqr = df[(df['Monthly_Salary'] < (Q1 - 1.5 * IQR)) | (df['Monthly_Salary'] > (Q3 + 1.5 * IQR))]
print("{:s}\n{:s}\n".format("Outliers of Monthly_salary found by IQR Method.", len("Outliers of Monthly_salary found by IQR Method.") * "-"))
outliers_iqr
```

No Outliers were found by IQR method in checking Monthly_Salary column

```python
Q1 = df['Employee_Age'].quantile(0.25)  # Calculate IQR of column Employee_Age
Q3 = df['Employee_Age'].quantile(0.75)
IQR = Q3 - Q1
print("Q1 =",Q1)
print("Q3 =",Q3)  

# Identify outliers
outliers_iqr = df[(df['Employee_Age'] < (Q1 - 1.5 * IQR)) | (df['Employee_Age'] > (Q3 + 1.5 * IQR))]
print("{:s}\n{:s}\n".format("Outliers of Employee_Age found by IQR Method.", len("Outliers of Employee_Age found by IQR Method.") * "-"))
outliers_iqr
```

While checking Employee_Age column,6 Outliers were found by IQR method.

### 2.Plot the chart with age and salary

```python
plt.figure(figsize=(8,5))
plt.bar(df.Employee_Age,df.Monthly_Salary)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Employee Age vs Monthly Salary")
plt.show()
```

From the plot we can see that employees aged 35 to 45 receive the highest salaries, with a peak around 40 years old. Salaries decrease for employees younger than 35 and older than 45.

## Data Encoding:

## Feature Scaling:

###  Perform the scaling of the features using standardscaler and minmaxscaler.

```python
from sklearn.preprocessing import StandardScaler

# Creating a scaler object
scaler = StandardScaler()

# Fit the scaler on the data
scaler.fit(df)

# Transform the data
scaled_data = scaler.transform(df)

# Print the scaled data
print(scaled_data.round(2))
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create a scaler object
scaler = MinMaxScaler()

# Fit the scaler on the data
scaler.fit(df)

# Transform the data
scaled_data = scaler.transform(df)

# Print the scaled data
print(scaled_data.round(1))
```
